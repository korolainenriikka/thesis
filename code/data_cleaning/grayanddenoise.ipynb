# Generate sample figure for preprocessing chapter

```python
import cv2
from torchvision.transforms import Grayscale, GaussianBlur
import matplotlib.pyplot as plt
import torch
from torchvision import transforms
from PIL import Image
```

```python
image_pil = Image.open('../../tex/images/noise.png')

# Create a Grayscale transform
to_grayscale = transforms.Grayscale(num_output_channels=1)
blur = transforms.GaussianBlur(3)

# Apply the transform
grayscale_image = to_grayscale(image_pil)
blurred = blur(grayscale_image)

```

```python
plt.imshow(blurred, cmap='gray')
blurred.save('../../tex/images/denoised.png')
```
